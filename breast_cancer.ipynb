```python
y=df['diagnosis']

# Encode categorical labels to numerical (0 and 1)
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
y = label_encoder.fit_transform(y)
```

```python
# splitting in trainig and test
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test= train_test_split(X,y,random_state=32)
```

```python
## feature scaling
from sklearn.preprocessing import StandardScaler
sc= StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
X_train
```

## Creating Neural network(ANN)


```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import ReLU,ELU,PReLU,LeakyReLU
from tensorflow.keras.layers import Dropout
```

*Sequential*

**a huge block of neurons in which forward and backward propagation will happen this can be done by sequential**

*Dense*

**to create neurons(hidden layer,input layer , output layer ) in our model we use  dense**

*Activation *

**tanh, relu , prelu usedd as activation fn which gives numerical output**

*Dropout*

** sometime model used to overfit so we use dropout (same as regularization in ML) it will deactivatae the neurons in our model to reduce overfitting**

```python
## Part 1
## initializing the ANN
classifier = Sequential()
```

```python
## adding input layer
classifier.add(Dense(units=30,activation='relu'))## relu applied to next upcoming layer
```

```python
## adding first hidden layer
classifier.add(Dense(units = 5 , activation = 'relu'))
#classifier.add(Dropout(0.1))
```

```python
## adding second hidden layer
classifier.add(Dense(units= 4,activation = 'relu'))
```

```python
## adding the output layer
classifier.add(Dense(1,activation= 'sigmoid'))
```

```python
# compling everything with adam optimizer and binary cross entropy as loss fn
classifier.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics=['accuracy'])
```

```python
# Early stopping(when accuracy does not increase significantly then model stops itself)
import tensorflow as tf
early_stopping = tf.keras.callbacks.EarlyStopping(
    monitor="val_loss",
    min_delta=0.1,
    patience=0.5,
    verbose=0.6,
    mode="auto",
    baseline=None,
    restore_best_weights=False,
    start_from_epoch=0,
)
```

```python
##Part2
## train our neural network
model_history  = classifier.fit(X_train,y_train,validation_split= 0.33,batch_size= 10,epochs = 100,callbacks= early_stopping)
```

```python
model_history.history.keys()
```

```python
##Part 3
## prediction for test data
y_pred = classifier.predict(X_test)
y_pred = (y_pred >= 0.5) # if output > 0.5 considering else rejecting
```

```python
## make confusion matrix
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test,y_pred)
cm
```

```python
# calculate accuracy
from sklearn.metrics import accuracy_score
score= accuracy_score(y_pred,y_test)
score
```
